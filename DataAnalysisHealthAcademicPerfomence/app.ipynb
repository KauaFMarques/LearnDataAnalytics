```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/student_habits_performance.csv")
```

```python
df
```

```python
cols=[
    "study_hours_per_day",
    "social_media_hours",
    "netflix_hours",
    "sleep_hours",
    "attendance_percentage",
    "exercise_frequency",
    "mental_health_rating",
    "exam_score"
]

sns.heatmap(df[cols].corr(),annot=True,cmap="coolwarm")
plt.title("correlaÃ§Ã£o entre os hÃ¡bitos e a nota final")
plt.show()
```

```python
sns.lmplot(data=df,x="study_hours_per_day",y="exam_score")
plt.title("Mais estudos->mMaiores Notas")
plt.show()
```

```python
filtro_estudo_alto= df["study_hours_per_day"]>5
filtro_estudo_baixo= df["study_hours_per_day"]<2

grupo_estudo_alto= df[filtro_estudo_alto]
grupo_estudo_baixo= df[filtro_estudo_baixo]

print("MÃ©dia de notas (estuda 5h):",filtro_estudo_alto.mean())
print("MÃ©dia de notas (estuda 2h):",filtro_estudo_baixo.mean())
```

```python
sns.histplot(data=df, x="social_media_hours")
x="social_media_hours"

plt.title("distribuiÃ§Ã£o de tempo em redes sociais")
plt.show()
```

```python
df["social_media_bin"]= pd.cut(
    df["social_media_hours"],
    bins=[0,2,4,6,10],
    labels=["0h-2h","2h-4h","4h-6h","+6h"]
)

sns.boxplot(x="social_media_hours",y="exam_score",data=df)
plt.title("Notas por tempo em redes sociais")
plt.show()
```

```python
for col in ["exercise_frequency", "mental_health_rating", "diet_quality"]:
    sns.boxplot(x=col, y="exam_score", data=df)
    plt.title(f"Notas por {col}") 
    plt.show()
```
